```python
import numpy as np
import pandas as pd

dataset = pd.read_csv('../datasets/Men T20I Player Innings Stats - 21st Century.csv')
```

```python
o_dataset = pd.read_csv('../datasets/T20-Edit - Sheet1.csv')

p_list = list(o_dataset.striker.unique())
p2_list = list(o_dataset.bowler.unique())

p_list  = set(p_list+p2_list)
pi_played = list(dataset['Innings Player'].unique())
common_p = list(set(p_list) & set(pi_played))
common_p
```

```python
dataset = dataset[dataset['Innings Player'].isin(common_p)]
```

# Calculating Batting Attributes

    -Number of Innings
    -Batting Average
    -Centuries
    -Fifties
    -Zeros
    -Highest Score
    -Avg. Strike Rate

```python
#calculate Number of innings played by the player

players_innings_batted = {}

unique_players = dataset['Innings Player'].unique()

#count the no of innings played
for i in unique_players:
    players_inn = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Innings Batted Flag'] == '1')][['Innings Batted Flag']].astype('int').sum().to_numpy()[0]
    players_innings_batted[i] = int(players_inn)
```

```python
#calculating batting average
#calculate - total runs scored / number of times dismissed 

batting_avg = {}

for i in unique_players:
    bat_out = len(dataset.loc[(dataset['Innings Player'] == i) & (dataset['Innings Not Out Flag'] == '0') ][['Innings Not Out Flag']])
    bat_runs_total = dataset.loc[(dataset['Innings Player'] == i)  & (dataset['Innings Runs Scored Num'] != '.')][['Innings Runs Scored Num']].astype('int').sum().to_numpy()[0]
    if bat_out != 0:
        batting_avg[i] = round(bat_runs_total/bat_out, 2)
    else:
        batting_avg[i] = 0
```

```python
#number of innings with centuries
century_innings = {}

for i in unique_players:
    count = dataset.loc[(dataset['Innings Player'] == i) & (dataset["Century"] == '1')][["Century"]].astype('int').sum().to_numpy()[0]
    century_innings[i] = count
```

```python
#number of innings with fifties
fifty_innings = {}

for i in unique_players:
    count = dataset.loc[(dataset['Innings Player'] == i) & (dataset["Fifty"] == '1')][["Fifty"]].astype('int').sum().to_numpy()[0]
    fifty_innings[i] = count
```

```python
#number of innings with zeros
zeros_innings = {}

for i in unique_players:
    count = len(dataset.loc[(dataset['Innings Player'] == i) & (dataset["Innings Runs Scored Num"] == '0')][["Innings Runs Scored Num"]])
    zeros_innings[i] = count
```

```python
#Highest score
highest_score = {}

for i in unique_players:
    score = dataset[(dataset['Innings Player'] == i) & (dataset['Innings Runs Scored Num'] != '.')]["Innings Runs Scored Num"].astype('int').max()
    highest_score[i] = score
```

```python
#Strike Rate
strike_rate_bat = {}

for i in unique_players:
    rate = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Innings Batting Strike Rate']!='.')][['Innings Batting Strike Rate']].astype('float').sum().to_numpy()[0]
    no_matches = len(dataset.loc[(dataset['Innings Player'] == i) & (dataset['Innings Batting Strike Rate']!='.')][['Innings Batting Strike Rate']])
    if no_matches == 0:
        strike_rate_bat[i] = 0
    else:
        strike_rate_bat[i] = round(rate/no_matches, 2)
```

# Calculating Bowling Attributes

    -Number of Innings
    -Overs
    -Bowling Average
    -More than Four Wickets
    -Strike Rate

```python
#calculate Number of innings played by the player
players_innings_bowled = {}

unique_players = dataset['Innings Player'].unique()

#count the no of innings played
for i in unique_players:
    players_inn = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Innings Bowled Flag'] == '1')][['Innings Bowled Flag']].astype('int').sum().to_numpy()[0]
    players_innings_bowled[i] = int(players_inn)
```

```python
# Number of overs Bowled
no_inn_bowled = {}

for i in unique_players:
    players_inn = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Innings Overs Bowled'] != '.') & (dataset['Innings Overs Bowled'] != 'DNB') & (dataset['Innings Overs Bowled'] != 'TDNB')][['Innings Overs Bowled']].astype('float').sum().to_numpy()[0]
    no_inn_bowled[i] = int(players_inn)
```

```python
#Bowling Average
#calculate - total runs conceded / number of wickets taken 

bowling_avg = {}

for i in unique_players:
    runs_conceded = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Innings Runs Conceded'] != '.') ][['Innings Runs Conceded']].astype('int').sum().to_numpy()[0]
    no_of_wick = dataset.loc[(dataset['Innings Player'] == i)  & (dataset['Innings Wickets Taken'] != '.')][['Innings Wickets Taken']].astype('int').sum().to_numpy()[0]
    if no_of_wick == 0:
        bowling_avg[i] = 0
    else:
        bowling_avg[i] = round(runs_conceded/no_of_wick, 2)
```

```python
#No of innings the player has taken >= Four Wickets

FF = {}

for i in unique_players:
    count = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Innings Wickets Taken'] != '.')][['Innings Wickets Taken']].astype('int')
    count = len(count.loc[count['Innings Wickets Taken']>=4])
    FF[i] = count
```

```python
#Strike Rate
strike_rate_bowl = {}

for i in unique_players:
    no_of_balls =  dataset.loc[(dataset['Innings Player'] == i) & ((dataset['Innings Overs Bowled']!='.') & (dataset['Innings Overs Bowled']!='DNB') & (dataset['Innings Overs Bowled']!='TDNB'))][['Innings Overs Bowled']].astype('float').sum().to_numpy()[0]
    temp = str(round(no_of_balls,2)).split('.')
    no_of_balls = (int(temp[0]) * 6) + int(temp[1])
    no_of_wickets = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Innings Wickets Taken']!='.')][['Innings Wickets Taken']].astype('int').sum().to_numpy()[0]
    if no_of_wickets == 0:
        strike_rate_bowl[i] = 0
    else:
        strike_rate_bowl[i] = round(no_of_balls/no_of_wickets,2)
```

# Calculating attributes against opposition
   
   
   
    -Number of Innings
    -Batting Average
    -Bowling Average
    -Centuries
    -Fifties
    -Zeros
    -Highest Score
    -Avg. Batting Strike Rate
    -Avg. Bowling Strike Rate
    -More than Four wickets
    
    
    Format:- 
             
       player  Op  NOI  Bat_Avg  Bowl_Avg  Centuries  Fifties  Zeros  Highest_Score  Avg_Bat_SR  Avg_Bowl_SR  FF
             
       ---------------------------------------------------------------------------------------------------------
       ---------------------------------------------------------------------------------------------------------
    
       ....

```python
#Number of Innings played by the batsman against opposition
alloppodata = []
# unique_teams = dataset.Opposition.unique()
for i in unique_players:
    unique_teams = dataset[dataset['Innings Player'] == i]['Opposition'].unique()
    for j in unique_teams:
        #no of innings
        batted_data = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset['Innings Batted Flag'] != '.')]['Innings Batted Flag'].astype('int').sum()
        bowled_data = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset['Innings Bowled Flag'] != '.')]['Innings Bowled Flag'].astype('int').sum()
        total_noi = batted_data + bowled_data
        
        #batting average
        bat_out = len(dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset['Innings Not Out Flag'] == '0') ][['Innings Not Out Flag']])
        bat_runs_total = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset['Innings Runs Scored Num'] != '.')][['Innings Runs Scored Num']].astype('int').sum().to_numpy()[0]
        bat_avg = 0
        if bat_out == 0:
            bat_avg = 0
        else:
            bat_avg = round(bat_runs_total/bat_out,2)
        
        #bowling average
        runs_conceded = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset['Innings Runs Conceded'] != '.') ][['Innings Runs Conceded']].astype('int').sum().to_numpy()[0]
        no_of_wick = dataset.loc[(dataset['Innings Player'] == i)  & (dataset['Opposition']==j) & (dataset['Innings Wickets Taken'] != '.')][['Innings Wickets Taken']].astype('int').sum().to_numpy()[0]
        bowl_avg = 0
        if no_of_wick == 0:
            bowl_avg = 0
        else:
            bowl_avg = round(runs_conceded/no_of_wick, 2)
            
        #Centuries
        century_ = dataset.loc[(dataset['Innings Player'] == i)  & (dataset['Opposition']==j) & (dataset["Century"] == '1')][["Century"]].astype('int').sum().to_numpy()[0]
        
        #Fifties
        fifty_ = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset["Fifty"] == '1')][["Fifty"]].astype('int').sum().to_numpy()[0]

        #Zeros
        zeros_ = len(dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset["Innings Runs Scored Num"] == '0')][["Innings Runs Scored Num"]])
        
        #highest score
        high_score = dataset[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset['Innings Runs Scored Num'] != '.')]["Innings Runs Scored Num"].astype('int').max()

        #Avg. Batting Strike Rate
        rate = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset['Innings Batting Strike Rate']!='.')][['Innings Batting Strike Rate']].astype('float').sum().to_numpy()[0]
        no_matches = len(dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset['Innings Batting Strike Rate']!='.')][['Innings Batting Strike Rate']])
        avg_strike_rate_bat = 0
        if no_matches == 0:
            avg_strike_rate_bat = 0
        else:
            avg_strike_rate_bat = round(rate/no_matches, 2)
            
        #Avg. Bowling Strike Rate
        no_of_balls =  dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & ((dataset['Innings Overs Bowled']!='.') & (dataset['Innings Overs Bowled']!='DNB') & (dataset['Innings Overs Bowled']!='TDNB'))][['Innings Overs Bowled']].astype('float').sum().to_numpy()[0]
        temp = str(round(no_of_balls,2)).split('.')
        no_of_balls = (int(temp[0]) * 6) + int(temp[1])
        no_of_wickets = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset['Innings Wickets Taken']!='.')][['Innings Wickets Taken']].astype('int').sum().to_numpy()[0]
        avg_strike_rate_bowl = 0
        if no_of_wickets == 0:
            avg_strike_rate_bowl = 0
        else:
            avg_strike_rate_bowl = round(no_of_balls/no_of_wickets,2)

        #FF
        FF_count = dataset.loc[(dataset['Innings Player'] == i) & (dataset['Opposition']==j) & (dataset['Innings Wickets Taken'] != '.')][['Innings Wickets Taken']].astype('int')
        FF_count = len(FF_count.loc[FF_count['Innings Wickets Taken']>=4])
    
        alloppodata.append([i, j, total_noi, bat_avg, bowl_avg, century_, fifty_, zeros_, high_score, avg_strike_rate_bat, avg_strike_rate_bowl, FF_count])
```

```python
against_oppo_df = pd.DataFrame(alloppodata, columns=["player",  "Op",  "NOI",  "Bat_Avg",  "Bowl_Avg",  "Centuries",  "Fifties",  "Zeros",  "Highest_Score",  "Avg_Bat_SR",  "Avg_Bowl_SR",  "FF"])
```

```python
against_oppo_df.to_csv('against_oppo_df.csv')
```

# Calculating Derived Attributes

```python
consistency = {}

for i in unique_players:
    bat_consis = round(0.4262*batting_avg[i] + 0.2566*players_innings_batted[i] + 0.1510*strike_rate_bat[i] + 0.0787*century_innings[i] + 0.0556*fifty_innings[i] - 0.0328*zeros_innings[i], 2)
    bowl_consis = round(0.4174*no_inn_bowled[i] + 0.2634*players_innings_bowled[i] + 0.1602*strike_rate_bowl[i] + 0.0975*bowling_avg[i] + 0.0615*FF[i],2)
    consistency[i] = [bat_consis, bowl_consis]
```

```python
import pickle

pickle_data = {'players_innings_batted': players_innings_batted, 'batting_avg':batting_avg, 'century_innings':century_innings,
              'fifty_innings':fifty_innings, 'zeros_innings':zeros_innings, 'highest_score':highest_score, 'strike_rate_bat':strike_rate_bat,
              'players_innings_bowled':players_innings_bowled, 'no_inn_bowled':no_inn_bowled, 'bowling_avg':bowling_avg,
              'FF':FF, 'strike_rate_bowl':strike_rate_bowl, 'against_oppo_df':against_oppo_df, 'consistency': consistency}

with open('player_performance.pkl', 'wb') as file:
    pickle.dump(pickle_data, file)
```
