```python
unique_match_ids = dataset.match_id.unique()

for i in unique_match_ids:
    unique_strikers = dataset[(dataset['match_id'] == i)]['striker'].unique()
    for j in unique_strikers:
        runs = dataset.loc[(dataset['match_id']==i) & (dataset['striker'] == j)]['runs_off_bat'].astype('int').sum()
        dataset.loc[(dataset['match_id']==i) & (dataset['striker'] == j), 'runs_individual'] = runs
```

```python
dataset = dataset.drop(['match_id', 'season', 'innings', 'ball', 'non_striker','bowler','runs_off_bat','extras', 'wicket_type','player_dismissed'],axis=1)
```

```python
teams = ['Australia', 'Ireland', 'Afghanistan', 'India', 'Zimbabwe', 
             'Scotland', 'Netherlands', 'Namibia', 'Kenya', 'Sri Lanka', 
             'Pakistan', 'England', 'South Africa', 'Bangladesh', 
             'West Indies', 'New Zealand', 'Bermuda']
```

```python
dataset = dataset[(dataset['batting_team'].isin(teams)) & (dataset['bowling_team'].isin(teams))]
```

```python
dataset
```

```python
team = dataset['batting_team'].unique()
striker = dataset['striker'].unique()

get_player_runs = {}

for i in striker:
    get_player_runs[i] = []
for i in striker:
    for j in team:
        data = list(dataset[(dataset['batting_team'] == j) & (dataset['striker'] == i)]['runs_individual'].unique())
        get_player_runs[i].append([data+[j]])
```
